1. What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
- Answer: In statistical hypothesis testing, we make decisions about a population based on sample data. Because we are relying on probabilities, there is always a risk of reaching an incorrect conclusion. These incorrect conclusions are classified into two specific types of errors: Type I and Type II errors.

A. Type I Error ($\alpha$ - Alpha):

Definition: A Type I error occurs when we reject the null hypothesis ($H_0$) when it is actually true. This is often referred to as a "False Positive.

Significance Level: The probability of committing a Type I error is denoted by the Greek letter alpha ($\alpha$). This is set by the researcher before the test (commonly at 0.05 or 5%).

Real-world Analogy: Convicting an innocent person. The null hypothesis is "Innocent," but the jury rejects it and decides "Guilty" incorrectly.

B. Type II Error ($\beta$ - Beta):

Definition: A Type II error occurs when we fail to reject the null hypothesis ($H_0$) when it is actually false. This is often referred to as a "False Negative.

Probability: The probability of committing a Type II error is denoted by beta ($\beta$).

Real-world Analogy: Acquitting a guilty person. The null hypothesis is "Innocent," and the jury fails to reject it, letting a guilty person go free.

- Impact on Decision-Making:

The Trade-off: There is an inverse relationship between the two errors. Reducing the risk of a Type I error (lowering $\alpha$) typically increases the risk of a Type II error ($\beta$), and vice versa, unless the sample size is increased.

Context Matters: Decision-makers must decide which error is more costly.
In medical testing (e.g., cancer screening), a Type II error (missing a cancer diagnosis) is often considered more dangerous than a Type I error (a false alarm). Therefore, tests are designed to minimize $\beta$.
In criminal justice, a Type I error (convicting the innocent) is viewed as worse, so the system requires proof "beyond a reasonable doubt" (minimizing $\alpha$).

Power of the Test: Decision-making quality is also measured by the Power of the Test ($1 - \beta$), which is the probability of correctly rejecting a false null hypothesis.



2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
- Answer: The P-value (probability value) is a fundamental concept in statistical hypothesis testing used to determine the significance of the results.

Definition: The P-value is the probability of observing test results at least as extreme as the results actually observed during the test, assuming that the null hypothesis ($H_0$) is correct.2.

Interpretation in Context of the Null Hypothesis: The P-value acts as a measure of evidence against the null hypothesis.
Low P-value (Strong Evidence against $H_0$): If the P-value is very small, it indicates that the observed data is highly unlikely to have occurred by random chance if the null hypothesis were true.
High P-value (Weak Evidence against $H_0$): If the P-value is large, the observed data is consistent with what we would expect under the null hypothesis.

Decision Rule: To make a decision, the P-value is compared to a pre-determined significance level ($\alpha$), usually set at 0.05 (5%). If $P \le \alpha$: We reject the null hypothesis. The result is statistically significant. If $P > \alpha$: We fail to reject the null hypothesis. The result is not statistically significant.

Common Misconception: The P-value is not the probability that the null hypothesis is true, nor is it the probability that the alternative hypothesis is false. It strictly measures the "weirdness" or "extremeness" of the sample data assuming $H_0$ is true.



3. Explain the difference between a Z-test and a T-test, including when to use each.
- Answer:
In the realm of inferential statistics, the Z-test and T-test are the two fundamental parametric methods used to determine if there is a significant difference between population means. While they share the same objective—evaluating hypotheses about a population based on sample data—they are distinct in their underlying statistical theories, assumptions, and application criteria. To answer this question comprehensively, we must explore the theoretical basis of each test, the critical role of the population standard deviation, the influence of sample size, and the specific decision rules that dictate when to use one over the other.

- Theoretical Basis and Statistical Distributions:

The primary difference between the two tests lies in the probability distribution they employ to calculate the critical region. The Z-test is based on the Standard Normal Distribution (also known as the Z-distribution), which is a bell-shaped curve with a mean of 0 and a standard deviation of 1. This distribution assumes that we have perfect knowledge of the population's variability. Because it relies on exact population parameters, the Z-test provides a precise determination of probability but is rigid in its requirements.

In contrast, the T-test relies on Student’s t-distribution. This distribution is similar to the normal distribution—symmetric and bell-shaped—but it has "heavier tails," meaning there is a higher probability of extreme values. The t-distribution is designed to account for the uncertainty introduced when we estimate the population’s variability using sample data. The shape of the t-distribution is not fixed; it changes based on the degrees of freedom (5$df = n - 1$). As the sample size increases, the t-distribution tightens and converges to the standard normal distribution, which explains why the two tests yield nearly identical results with very large sample sizes.

- The Critical Factor: Population Standard Deviation ($\sigma$):

The most decisive factor in choosing between a Z-test and a T-test is the knowledge of the population standard deviation ($\sigma$). This parameter represents the true spread or variability of the entire population.

A Z-test is strictly used when the population standard deviation is known. This is a rare occurrence in practical research, usually reserved for standardized manufacturing processes or census data where historical records provide an exact $\sigma$. When $\sigma$ is known, the researcher can standardize the sample mean directly into a Z-score using the formula:$$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$$Here, the denominator ($\sigma / \sqrt{n}$) is the standard error calculated from the true population parameter.

A T-test is used when the population standard deviation is unknown, which is the case for the vast majority of scientific and social research. Since 9$\sigma$ is unavailable, the researcher must estimate it using the sample standard deviation (10$s$). This estimation introduces additional error/uncertainty into the test statistic. To penalize for this uncertainty, the T-test uses a slightly different formula:$$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$$The use of $s$ instead of $\sigma$ necessitates the use of the t-distribution (and its degrees of freedom) to determine statistical significance, providing a more conservative test that is less likely to produce a Type I error (false positive) under uncertainty.

- The Role of Sample Size ($n$):

While the population standard deviation is the theoretical differentiator, sample size dictates the practical application of these tests, governed by the Central Limit Theorem (CLT).

The Z-test is typically associated with large sample sizes (generally defined as 14$n \geq 30$). According to the Central Limit Theorem, as the sample size becomes large, the sampling distribution of the mean approximates normality regardless of the population's original distribution. Consequently, if a sample is large ($n > 30$), some statisticians argue that a Z-test can be used as an approximation even if $\sigma$ is unknown, because the sample standard deviation ($s$) becomes a highly accurate estimator of the population standard deviation ($\sigma$).

The T-test is the mandatory choice for small sample sizes (17$n < 30$). When sample sizes are small, the sampling distribution may not be perfectly normal, and the estimate of the standard deviation (19$s$) can be volatile and inaccurate. The T-test’s heavier tails compensate for this volatility. However, modern statistical practice suggests that the T-test is valid and preferred for any sample size if $\sigma$ is unknown. Using a T-test on a large sample is perfectly acceptable (and increasingly standard in software like SPSS or Python) because the t-values naturally align with Z-values as $n$ grows infinite.

- Assumptions and Degrees of Freedom:

Both tests operate under specific assumptions that must be met to ensure validity. Both assume that the data is continuous, the samples are randomly selected, and the observations are independent. However, the normality assumption is stricter for the T-test when 23$n$ is small; the data must come from a population that is approximately normally distributed. If the data is heavily skewed and $n$ is small, non-parametric alternatives (like the Mann-Whitney U test) should be considered instead.

A unique feature of the T-test is the concept of degrees of freedom (df).25 In a Z-test, the critical value (e.g., 1.96 for a 95% confidence level) is a fixed constant. In a T-test, the critical value changes depending on the sample size. For example, with a sample size of 10 ($df=9$), the critical value for a 95% confidence interval is 2.262, which is much higher than 1.96. This wider interval reflects the "penalty" for having less data, making it harder to reject the null hypothesis compared to a Z-test.

Summary of Application (When to Use Which):
To summarize the decision-making process for this question, one should apply the following logic:

Use the Z-test when: Your sample size is large (27$n \geq 30$). You possess the true population standard deviation ($\sigma$).You are comparing a sample mean to a population mean (One-Sample Z-Test) or comparing two proportions.

Use the T-test when: You do not know the population standard deviation (you only have 29$s$). Your sample size is small (31$n < 30$), although it is also statistically valid for large samples if 32$\sigma$ is unknown. You are comparing two independent groups (Independent Samples T-Test) or related groups (Paired Samples T-Test) in experimental settings.

In conclusion, while the Z-test serves as an idealized model for large samples with known parameters, the T-test is the workhorse of practical statistics, offering the necessary robustness to handle the uncertainty inherent in estimating population variance from limited sample data.



4. What is a confidence interval, and how does the margin of error influence its width and interpretation?
- Answer:

A. Confidence Interval (CI):A Confidence Interval is a range of values, derived from sample statistics, that is likely to contain the true population parameter (such as the mean) with a specified level of confidence (e.g., 95% or 99%).

It is calculated as: $\text{Point Estimate} \pm \text{Margin of Error}$.

A 95% CI means that if we repeated the sampling process infinite times, 95% of the calculated intervals would contain the true population mean.

B. Margin of Error (ME):The Margin of Error expresses the maximum expected difference between the true population parameter and the sample estimate. It accounts for sampling error.Formula (for Z-test): $ME = Z \times \frac{\sigma}{\sqrt{n}}$

C. Influence on Width and Interpretation: The Margin of Error directly dictates the width (precision) of the Confidence Interval.
- Larger Margin of Error:

Result: The CI becomes wider.Interpretation: We are less precise about the location of the true parameter. A wide interval (e.g., 10 to 90) gives very little specific information.

- Smaller Margin of Error:

Result: The CI becomes narrower.

Interpretation: We have higher precision. A narrow interval (e.g., 48 to 52) provides a very specific estimate of the population parameter.

Factors affecting ME:

Sample Size ($n$): Increasing $n$ decreases the ME (narrower interval).

Confidence Level: Increasing confidence (e.g., 95% to 99%) increases the Z-score, which increases the ME (wider interval).

Variability ($\sigma$): Higher data variability increases the ME (wider interval).

5. Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
- Answer:

A. Purpose of ANOVA (Analysis of Variance):

ANOVA is a statistical method used to compare the means of three or more independent groups to determine if at least one group mean is statistically different from the others.

Null Hypothesis ($H_0$): $\mu_1 = \mu_2 = \mu_3 = ... = \mu_k$ (All group means are equal).

Alternative Hypothesis ($H_1$): At least one mean is different.

B. Assumptions: For ANOVA results to be valid, three key assumptions must be met:

Normality: The data in each group should follow a normal distribution.

Homogeneity of Variances (Homoscedasticity): The variances within each group should be approximately equal.

Independence: Observations must be independent of each other.

C. Extension of Hypothesis Testing: One might ask, "Why not just perform multiple T-tests?" (e.g., Group A vs B, B vs C, A vs C).

The Problem: Conducting multiple pairwise comparisons increases the Type I error rate exponentially. If you run multiple tests at $\alpha = 0.05$, the probability of finding a false positive purely by chance accumulates.

The Solution: ANOVA compares all groups simultaneously using the F-statistic. The F-statistic compares the ratio of Between-Group Variability (differences caused by the treatment/factor) to Within-Group Variability (random error).If Between-Group variability is significantly higher than Within-Group variability, we conclude that the groups are distinct.

```python
# 6. Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

"Answer: This solution uses statsmodels to perform a Z-test. We test if the sample mean differs significantly from a hypothesized population mean."

import numpy as np
from statsmodels.stats.weightstats import ztest

# 1. Define the Dataset
# Let's assume this data represents the weights of a product sample (n > 30)
data = [102, 101, 99, 100, 103, 98, 101, 104, 100, 99,
        102, 101, 97, 103, 100, 101, 99, 102, 105, 100,
        101, 102, 98, 99, 103, 100, 101, 104, 99, 102, 100]

# 2. Define Hypotheses
# H0: Population Mean = 100
# H1: Population Mean != 100
population_mean_h0 = 100
alpha = 0.05

# 3. Perform Z-test
# value argument represents the mean under the Null Hypothesis
z_score, p_value = ztest(data, value=population_mean_h0)

# 4. Output and Interpretation
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

print("\n--- Interpretation ---")
if p_value < alpha:
    print("Reject the Null Hypothesis.")
    print("There is significant evidence that the mean is different from 100.")
else:
    print("Fail to reject the Null Hypothesis.")
    print("There is no significant evidence that the mean differs from 100.")
```

```python
# 8. Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

"Answer: The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean will approximate a normal distribution as the sample size increases, even if the original population is not normal."

import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Non-Normal Population (Exponential Distribution)
population_data = np.random.exponential(scale=1.0, size=10000)

# 2. Implement CLT: Take many samples and calculate their means
sample_means = []
sample_size = 50   # n > 30 is ideal for CLT
num_samples = 1000

for _ in range(num_samples):
    # Randomly sample 'n' elements from the population
    sample = np.random.choice(population_data, size=sample_size)
    sample_means.append(np.mean(sample))

# 3. Visualization
plt.figure(figsize=(12, 5))

# Plot 1: Original Non-Normal Distribution
plt.subplot(1, 2, 1)
plt.hist(population_data, bins=50, color='orange', alpha=0.7)
plt.title('Original Population (Exponential)')
plt.xlabel('Value')

# Plot 2: Sampling Distribution of the Mean
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title(f'Sampling Distribution of Means (n={sample_size})')
plt.xlabel('Sample Mean')

plt.tight_layout()
plt.show()
```

```python
# 9. Write a Python function to calculate and visualize the confidence interval for a sample mean.

"Answer: This function calculates the 95% confidence interval using the T-distribution (appropriate for general samples) and visualizes it using error bars."

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_confidence_interval(data, confidence=0.95):
    # Calculate statistics
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard Error of the Mean

    # Calculate Margin of Error using T-distribution
    margin_of_error = sem * stats.t.ppf((1 + confidence) / 2., n-1)

    start = mean - margin_of_error
    end = mean + margin_of_error

    print(f"Mean: {mean:.2f}")
    print(f"95% Confidence Interval: [{start:.2f}, {end:.2f}]")

    # Visualization
    plt.figure(figsize=(6, 4))
    plt.errorbar(x=1, y=mean, yerr=margin_of_error, fmt='o',
                 color='blue', ecolor='red', capsize=10, label='Mean ± ME')
    plt.title(f'{int(confidence*100)}% Confidence Interval')
    plt.xticks([]) # Hide x-axis ticks
    plt.ylabel('Value')
    plt.ylim(min(data)-1, max(data)+1)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Example Usage
sample_data = [12, 15, 14, 10, 13, 15, 14, 18, 12, 13]
visualize_confidence_interval(sample_data)
```

```python
# 10. Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

"Answer: A Chi-square goodness-of-fit test checks if categorical data follows a hypothesized frequency distribution. Scenario: A die is rolled 60 times. We expect a fair die to show each face 10 times. We test if the observed rolls differ significantly from a fair die."

import scipy.stats as stats

# 1. Data Setup
# Observed frequencies (What actually happened)
observed = [8, 13, 11, 7, 12, 9]

# Expected frequencies (Theoretical - Fair die: 60 rolls / 6 faces = 10 each)
expected = [10, 10, 10, 10, 10, 10]

# 2. Perform Chi-Square Test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# 3. Interpretation
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject Null Hypothesis (The die is likely biased).")
else:
    print("Result: Fail to reject Null Hypothesis (The die is likely fair).")
```
